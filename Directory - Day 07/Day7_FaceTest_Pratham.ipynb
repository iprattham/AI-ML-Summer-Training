```python
#load any sample for the dataset
#ask the coder - user id and select any random 1-10 sample number of that user id
#as the query image

import matplotlib.image as mimg
import matplotlib.pyplot as plt
import numpy as np

# load the query image
usrId = int(input("Enter the User number: "))
samp = np.random.randint(1,10) #selcting any random sample number
path = "/content/drive/MyDrive/Colab Notebooks/orl_face/u%d/%d.png"%(usrId,samp)

#read the img
img = mimg.imread(path)

#plot the img
print(type(img))
print(img.shape)

#display the img
plt.figure(1,(5,5))
plt.imshow(img, cmap='gray')
plt.axis('off')
plt.title("Query Image")
```

```python
import joblib
from sklearn import svm ,metrics
import numpy as np
from sklearn.metrics import mean_squared_error

# calcualte the features of the query image
feat_query = img.reshape(1,-1)

# model path
path_model = '/content/drive/MyDrive/Colab Notebooks/orl_face_trained_model.pkl'
# load the trained model
face_model = joblib.load(path_model)


# predict the id of the query image
id = face_model.predict(feat_query)
print("User id predicted is :",int(id[0]))


path = "/content/drive/MyDrive/Colab Notebooks/orl_face/u%d/%d.png"%(id[0],1)

im_pred = mimg.imread(path)

plt.figure(1,(8,5))
plt.subplot(1,2,1)
plt.imshow(img,cmap='gray')
plt.axis('off')
plt.title("Query image")

plt.subplot(1,2,2)
plt.imshow(im_pred,cmap='gray')
plt.axis('off')
plt.title("Predicted user")

# Assuming 'img' and 'im_pred' are grayscale images of the same size
mse = mean_squared_error(img, im_pred)
percentage_match = (1 - mse) * 100  # Higher percentage indicates better match

print(f"\nPercentage match: {percentage_match:.2f}%")
```

```python
import matplotlib.image as mimg
import matplotlib.pyplot as plt
import numpy as np

total_samples = 61
# Initialize data and images as lists to accommodate varying image sizes
data = []
label = np.zeros((total_samples))
images = []

label[40:] = 1

print(label.shape)

# Determine the size of the first image to initialize the data array
first_image_path = "/content/drive/MyDrive/Colab Notebooks/brain_tumor/brain_tumor/d(1).png"
first_image = mimg.imread(first_image_path)
first_image_shape = first_image.shape

# Initialize data array with the correct dimensions
data = np.zeros((total_samples, first_image_shape[0] * first_image_shape[1]))

for i in range(1,total_samples+1,1):
  path = "/content/drive/MyDrive/Colab Notebooks/brain_tumor/brain_tumor/d(%d).png"%(i)
  im = mimg.imread(path)

  # Handle potential color channels
  if len(im.shape) == 3:
    temp_im = im[:,:,0] # Extract a single channel if the image has multiple
  else:
    temp_im = im # Keep the image as is if it's grayscale

  # Resize the image to match the first image
  temp_im = np.resize(temp_im, (first_image_shape[0], first_image_shape[1]))

  feat = temp_im.reshape(1,-1)

  data[i-1,:] = feat # Assign the reshaped image data to the array
  images.append(temp_im) # Append the image to the list

print(label)

# Convert images to a numpy array
images = np.stack(images, axis=0)

print(data.shape)
print(images.shape)
```

```python
from sklearn import svm, metrics, model_selection
import numpy as np
import pandas as pd

x = data
y = label
xtrain,xtest,ytrain,ytest = model_selection.train_test_split(x,y,test_size=0.3,random_state=4)

tumor_model = svm.SVC(kernel='linear', gamma="scale", C=2)
tumor_model.fit(xtrain,ytrain)

ypred = tumor_model.predict(xtest)

kernel = ["poly", "rbf", "linear"]
Cvalue = [1,2,3]

result = np.zeros((len(kernel),len(Cvalue)))
for i in range(len(kernel)):
  for j in range(len(Cvalue)):
    tumor_model = svm.SVC(kernel=kernel[i], gamma="scale",C=Cvalue[i])
    tumor_model = tumor_model.fit(xtrain,ytrain)
    ypred = tumor_model.predict(xtest)

    acc = metrics.accuracy_score(ypred,ytest)
    result[i,j]=acc

ResultDF = pd.DataFrame(result,index=kernel,columns=["C=1","C=2","C=3"])
ResultDF = ResultDF.T
print(ResultDF)
ResultDF.plot(kind='bar',figsize=(4,4))
```
