```python
# performing some arithmetic operations manually

# accessing the scores of all games
score = data['score']
#average of the score
avg = score.mean()
#minimum of the score
mn = score.min()
#maximum of the score
mx = score.max()
#variance of the score
var = score.var()
#standard deviation of the score
std = score.std()

print("Average is: ",avg)
print("Variance is: ",var)
print("Minimum is: ",mn)
print("Maximum is: ",mx)
print("Standard Deviation is: ",std)
```
