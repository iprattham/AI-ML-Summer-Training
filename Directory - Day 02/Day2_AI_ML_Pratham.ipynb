**# MATRIX CREATION**

```python
import numpy as np
```

WAP to create an autothreshhold algorithm.
1. create a 9x9 matrix within a range of 0-100 integer number.
2. find the avg of each row and save it in a B list (1x9).

Step 2: [B0,B,B2,B3,B4,B5,B6,B7,B8]
3. Now, Compare the values of each row with its respective MEAN. If mean vcalue is greater than the element, then leave the element as it is, else make the element 0. Create a new matric C to hold the updated result.
4. Count the no. of zeros & non zero elements, which should be equal to the no. of elements in a matrix.

```python
# creating a 9x9 matrix
np.random.seed(5)
A = np.random.randint(0,100,size=(9,9))
print("Input Matrix")
print(A.shape[0])
print ("\n",A)

#step 2
B = [] #empty list for storing means

for i in range (0,A.shape[0],1):
    temp = A[i,:].mean()
    B.append(temp) #save the mean on the respective position of the B list
print("\n",B)

#step 3 & 4
C = np.zeros((A.shape[0],A.shape[1]))
ZerosCount = 0
NonZerosCount = 0
for i in range(A.shape[0]):
    for j in range(A.shape[0]):
        if (A[i,j])>B[i]:
            C[i,j] = 0
            ZerosCount=ZerosCount+1
        else:
            C[i,j] = A[i,j]
            NonZerosCount=NonZerosCount+1

print("\nNo. of zeros are: ",ZerosCount)
print("No. of non-zeros are: ",NonZerosCount)

if (NonZerosCount+ZerosCount == A.size):
    print("\nUpdated Matrix is Verified!")
else:
    print("\nUpdated Matrix is not validated!")

print("\nUpdated Matrix")
print(C)
```
